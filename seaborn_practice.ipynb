```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r"deliveries.csv")
df
```

```python
# 1. Find ans visualize the total runs scored by each team.

a = df.groupby('batting_team')['total_runs'].sum()
print(a)
color = sns.color_palette("rainbow", len(a))
sns.barplot(x = a.index, y = a.values)
plt.xticks(rotation = 90)
```

```python
#2.Identify and display the names of the top 10 highest-scoring batters
# with their total runs scored.

a = df.groupby('batter')['total_runs'].sum().sort_values(ascending = False).head(10)
sns.barplot(x = a.index, y = a.values, palette="magma")
plt.xticks(rotation = 10)
```

```python
#3. Find and visualize the top 10 bowlers with the most wickets

valid_wickets = df[df["is_wicket"] == 1]
valid_wickets = valid_wickets[~valid_wickets["dismissal_kind"].isin(["run out", "retired hurt"])]


bowler_wickets = valid_wickets["bowler"].value_counts().reset_index()
bowler_wickets.columns = ["Bowler", "Wickets"]


top_10_bowlers = bowler_wickets.head(10)
print(top_10_bowlers)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_bowlers["Wickets"], y=top_10_bowlers["Bowler"], palette="crest")
plt.xlabel("Total Wickets")
plt.ylabel("Bowler")
plt.title("Top 10 Bowlers with Most Wickets in IPL")
plt.show()
```

```python
#4. Tabulate and visualize the total number of extras conceded by each team.

data = df.groupby('bowling_team')['extra_runs'].sum()
print(data)
sns.barplot(x = data.values, y = data.index, palette="flare")
```

```python
#5. Tabulate and visualize the batsmen (top 5) involved in the most run-outs.
d = df[df['dismissal_kind'] == 'run out']
ab = d['player_dismissed'].value_counts().head()
print(ab)
sns.barplot(y = ab.values, x = ab.index, palette="mako")
plt.xticks(rotation = 90)
```

```python
#6. Tabulate and visualize the top 5 bowlers who conceded the most extras.

f = df.groupby('bowler')['extra_runs'].sum().sort_values(ascending = False).head()
print(f)
sns.barplot(y = f.values, x = f.index, palette = "Spectral")
plt.xticks(rotation = 90)
```

```python
#7. Tabulate and visualize the most common dismissal type.
n = df['dismissal_kind'].value_counts()
print(n)
sns.barplot(x = n.values, y = n.index, palette = "viridis_r")
```

```python
 #8. Find and display which of the top 3 overs has the highest run rate.

v = df.groupby(['match_id', 'batting_team', 'over'])['total_runs'].sum().sort_values(ascending=False).head(3)
v = v.reset_index()
v["runrate"] = v["total_runs"] / 6  
print(v)
# Plot using Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x=v["over"], y=v["runrate"], hue=v["batting_team"], palette="magma_r")
plt.xlabel("Over Number")
plt.ylabel("Run Rate")
plt.title("Top 3 Overs with Highest Run Rate")
plt.show()
```

```python
#9. Analyze and visualize powerplay performance (first 6 overs).
powerplay_df = df[(df['over'] >= 1) & (df['over'] <= 6)]

# Group by match and team to get total runs and wickets
grouped_df = powerplay_df.groupby(['match_id', 'inning', 'batting_team'])\
    .agg(total_runs=('total_runs', 'sum'), wickets=('player_dismissed', 'count'))\
    .reset_index()

# Plot total runs in powerplay
plt.figure(figsize=(12, 6))
sns.boxplot(x='batting_team', y='total_runs', data=grouped_df)
plt.xticks(rotation=90)

plt.show()

# Plot wickets lost in powerplay
plt.figure(figsize=(12, 6))
sns.boxplot(x='batting_team', y='wickets', data=grouped_df)
plt.xticks(rotation=90)
```

```python
#10.Tabulate and visualize the top 5 batsmen with the most sixes and fours.

sixes = df[df['batsman_runs'] == 6].groupby('batter').size().reset_index(name='sixes')
fours = df[df['batsman_runs'] == 4].groupby('batter').size().reset_index(name='fours')

# Merge the data
batsmen_stats = pd.merge(sixes, fours, on='batter', how='outer').fillna(0)

# Get top 5 batsmen based on sixes and fours combined
batsmen_stats['total_boundaries'] = batsmen_stats['sixes'] + batsmen_stats['fours']
top_batsmen = batsmen_stats.sort_values(by='total_boundaries', ascending=False).head(5)
print(top_batsmen)
# Plot the data
plt.figure(figsize=(10, 6))
sns.barplot(data=top_batsmen.melt(id_vars=['batter'], value_vars=['sixes', 'fours']),
            x='batter', y='value', hue='variable', palette=['red', 'blue'])
plt.xticks(rotation = 90)
```
